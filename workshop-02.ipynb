```python
import time
import psutil
from datetime import datetime
import csv

CSV_FILENAME = "workshop2.csv"

def write_header(header_row):
    f = open(CSV_FILENAME, "w")
    csv_writer = csv.writer(f)
    csv_writer.writerow(header_row)
    f.close()

def append_to_csv(row):
    f = open(CSV_FILENAME, "a")
    csv_writer = csv.writer(f)
    csv_writer.writerow(row)
    f.close()

def collect_metrics():
    print( " I am collecting cpu and mem metrics ")
    #create two variable cpu, mem inside this function 
    # assign some dummy values
    cpu = psutil.cpu_percent()
    mem = psutil.virtual_memory().percent
    timestamp = str(datetime.now())
    
    # String formatting ?
    #print (  " Cpu = {} , Mem = {} , TimeStamp = {} ".format( cpu, mem, timestamp  )   )
    return (cpu, mem, timestamp)


write_header( ["cpu", "mem", "timestamp"]  )
while True:
    row = collect_metrics()
    append_to_csv(row)
    time.sleep(30)
```
